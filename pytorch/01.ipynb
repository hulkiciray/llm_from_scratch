### squeeze & unsqueeze
If you are having trouble understanding what `squeeze` and `unsqueeze` do in pytorch, here is an explanation:

The unsqueeze function in PyTorch is used to add an extra dimension to a tensor at a specified position. This new dimension has a size of 1, often referred to as a singleton dimension. Adding such dimensions can be important when you need to align tensor shapes for operations that require specific dimensionality.

````python
import torch
from IPython.display import Image
````

````python
tensor = torch.tensor([[1,2,3], [4,5,6]])
print(f"tensor:\n {tensor}")
print(f"tensor shape:\n {tensor.shape}")
````

````python
print(f"unsqueeze:\n {torch.unsqueeze(tensor, dim=0)}")
print(f"shape after unsqueeze:\n {torch.unsqueeze(tensor, dim=0).shape}")
````

````python
print(f"unsqueeze:\n {torch.unsqueeze(tensor, dim=1)}")
print(f"shape after unsqueeze:\n {torch.unsqueeze(tensor, dim=1).shape}")
````

````python
print(f"unsqueeze:\n {torch.unsqueeze(tensor, dim=-1)}")
print(f"shape after unsqueeze:\n {torch.unsqueeze(tensor, dim=-1).shape}")
````

### torch.flatten
Flattens input by reshaping it into a one-dimensional tensor. If `start_dim` or `end_dim` are passed, only dimensions starting with `start_dim` and ending with `end_dim` are flattened. The order of elements in input is unchanged.

reference: https://pytorch.org/docs/stable/generated/torch.flatten.html

````python
print(f"Tensor: {tensor}\n"
      f"shape of tensor: {tensor.shape}\n"
      f"Flatten: {torch.flatten(tensor)}, shape: {torch.flatten(tensor).shape}\n\n"

      f"Flatten with start_dim=1: \n{torch.flatten(tensor, start_dim=1)}, \nshape: {torch.flatten(tensor, start_dim=1).shape}\n"
      f"Flatten with start_dim=0: \n{torch.flatten(tensor, start_dim=0)}, \nshape: {torch.flatten(tensor, start_dim=0).shape}\n"
      f"Flatten with start_dim=-1: \n{torch.flatten(tensor, start_dim=-1)}, \nshape: {torch.flatten(tensor, start_dim=-1).shape}\n\n"

      f"Flatten with end_dim=1: \n{torch.flatten(tensor, end_dim=1)}, \nshape: {torch.flatten(tensor, end_dim=1).shape}\n"
      f"Flatten with end_dim=0: \n{torch.flatten(tensor, end_dim=0)}, \nshape: {torch.flatten(tensor, end_dim=0).shape}\n"
      f"Flatten with end_dim=-1: \n{torch.flatten(tensor, end_dim=-1)}, \nshape: {torch.flatten(tensor, end_dim=-1).shape}")
````

### torch.long

reference: https://stackoverflow.com/questions/67287559/pytorch-tensor-declared-as-torch-long-becomes-torch-int64

````python
import torch
a = torch.tensor([[0, 1, 1, 2],
                  [1, 0, 2, 1]], dtype=torch.long)
a.dtype
````
